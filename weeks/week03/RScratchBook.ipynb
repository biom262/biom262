```r
hot<-read.table("hot.txt",header=T)
```

```r
hotminus<-subset(hot, hot$hotsix=="-")
hotplus<-subset(hot,hot$hotsix=="+")
shapiro.test(hotminus$lukewarm-hotplus$lukewarm)
shapiro.test(hotminus$tepid-hotplus$tepid)
shapiro.test(hotminus$mild1-hotplus$mild1)
shapiro.test(hotminus$athermal-hotplus$athermal)
shapiro.test(hotminus$bathwater-hotplus$bathwater)
shapiro.test(hotminus$coldshower-hotplus$coldshower)
```
