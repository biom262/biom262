```r
lm<-read.table("littermates.txt",head=T)

plot(lm)
plot(mut~ctl,ylim=c(0,.6),xlim=c(0,.6),col="red")
abline(0,1, lty=2, col="salmon")
```

```r
#How do the variances of the two samples compare? P-values are close to 1; pretty similar variance
var.test(mut,ctl)
```

```r
#Shapiro test; test for normality; smaller p-value is less normal
shapiro.test(ctl-mut)
```

```r
#Since they are normally distributed, can do t-test. Need to tell if paired or unpaired.
t.test(mut,ctl,paired=T)
```

```r
#What if we unpaired the samples?
t.test(mut,ctl)
```
